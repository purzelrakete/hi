# Logistic Regression in numpy

Implement logistic regression using numpy.

### Generate gaussian mixture

First, let's generate some data. Well use a gaussian mixture model to generate 2 classes in 2 dimensions. We should be able to fit a model and then separate those two classes.

```python
m, n, nclasses, bounds = 200, 2, 2, 5
X, y = data.gaussian_clusters(nclasses, m, n, bounds)

plt.figure(figsize = (PLOT_WIDTH, PLOT_HEIGHT))
plt.scatter(X[0, :], X[1, :], c = y, cmap = plt.cm.seismic);
```

### Fit and plot decision boundary

```python
model = logreg.train(X, y)
```

```python
bfrom, bto = [-bounds, model.boundary(-bounds)], [bounds, model.boundary(bounds)]

plt.figure(figsize = (PLOT_WIDTH, PLOT_HEIGHT))
plt.scatter(X[0, :], X[1, :], c = y, cmap = plt.cm.seismic)
plt.plot([bfrom[0], bto[0]], [bfrom[1], bto[1]]);
```
